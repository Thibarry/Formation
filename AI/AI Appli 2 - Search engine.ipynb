```python
df = pd.read_csv(f"{Input_path}/books_metadata.csv")
display(df)
```

```python
from nltk import word_tokenize

kys = df["Keywords"].values
b1_kys = word_tokenize(kys[0])
b1_kys.remove(',')
b1_kys
```

```python
import constraint
# Creating Combinations
com = constraint.Problem()
com.addVariables(["a","b"], b1_kys)
com.addConstraint(constraint.AllDifferentConstraint())
print(com.getSolutions())
```

```python
wrds = []
for ele in com.getSolutions():
    wrds.append(f"{ele['a']} "+f"{ele['b']}")
    
wrds
```

```python
Titles = df['Title'].values
Kys = df["Keywords"].values
keywrds = {'B1': [], 'B2': [], 'B3': [], 'B4': [],
           'B5': [], 'B6': [], 'B7': []}
kys = []
wrds = []
for i in range(7):
    kys.clear()
    wrds.clear()
    kys = word_tokenize(Kys[i])
    while (kys.count(',')):
        kys.remove(',')
    com = constraint.Problem()
    com.addVariables(["a","b"], kys)
    com.addConstraint(constraint.AllDifferentConstraint())
    for ele in com.getSolutions():
        wrds.append(f"{ele['a']} "+f"{ele['b']}")
        
    for k in wrds:
        keywrds[f"B{i+1}"].append(k)
```

```python
keywrds
```

```python
Df = []
for bk in list(keywrds.keys()):
    for ks in keywrds[bk]:
        Df.append([ks,bk])
```

```python
import numpy as np
Df = np.array(Df)
KYS = pd.DataFrame({"keywrds":Df[:,0], "book":Df[:,1]})
KYS
```

```python
inp = Df[:,0]
y = Df[:,1]

enc = CountVectorizer()
enc.fit(inp)
e_inp = enc.transform(inp)

tf = TfidfTransformer()
tf.fit(e_inp)
X = tf.transform(e_inp)
```

```python
CModel = DecisionTreeClassifier()
CModel.fit(X,y)
```

```python
def search_book():
    query = input('Search: ')
    enc_q = enc.transform([query])
    tf_q = tf.transform(enc_q)
    bk = CModel.predict(tf_q)
    i = {'B1': 0, 'B2': 1, 'B3': 2, 'B4': 3,
        'B5': 4, 'B6': 5, 'B7': 6}
    title = df.values[i[bk[0]],0]
    desc = df.values[i[bk[0]],1]
    price = df.values[i[bk[0]],3]
    result = f"""
Tittle:
{title}

Description:
{desc}

Price:{price}"""
    print(result)
```
